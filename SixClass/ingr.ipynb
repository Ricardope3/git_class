```python
def mean(lista):
    compound = 0
    length = 0
    for elem in lista:
        if not pd.isna(elem):
            compound += elem
            length += 1
    if length == 0:
            return 0
    return compound/length
```

```python
dfs = []
debug = False
for i ,j in ing.iterrows():
    comparedTo = []
    likenessIndex = []
#     print(i)
    for ii ,jj in ing.iterrows():
        comparedTo.append(ii)
        likeness = []
#         print(ii)
        for columnName, columnValues in ing.iteritems():
            nominator = min(columnValues[ii],columnValues[i])
            denominator = max(columnValues[i],columnValues[ii])
#             print(nominator,denominator)
            if not isinstance(nominator,str):
                difference = nominator/denominator
            likeness.append(difference)
        likenessIndex.append(mean(likeness))

    data = {'Compared To':comparedTo, 'Likeness Index':likenessIndex}
    df = pd.DataFrame(data)
    if debug:
        print(df)
    dfs.append(df)
    
```
