```python
#haciendo una lista de los nutrientes de las columnas
nutrients = ing.columns.values.tolist ()
print(nutrients)
```

```python
#lista de índices(ingredientes) para comparar
ingredient_list = []
for ingrediente, _ in ing.iterrows():
    ingredient_list.append(ingrediente)
ingredient_list
```

```python
#diccionario principal donde están los diccionarios de cada ingrediente
ingr_dict = {}
for ingr in ingredient_list: 
    ingr_dict[ingr] = {}
print(ingr_dict)
```

```python
#lista de listas de los valores de nutrientes de cada ingrediente. aka: every row saved in a different list
value_list = []
for ingrediente, _ in ing.iterrows():
    temp_nutr_list = []
    for _, columnValues in ing.iteritems():
        nutriente = columnValues[ingrediente]
        temp_nutr_list.append(nutriente)
    value_list.append(temp_nutr_list)
#print(value_list)
  
            
```

```python
#lista provisional de todos los valores de value list en una única lista
complete_value_list = []
for ingrediente, _ in ing.iterrows():
    for _, columnValues in ing.iteritems():
        nutriente = columnValues[ingrediente]
        complete_value_list.append(nutriente)
#print(complete_value_list)
```

```python
#introduciendo la lista de nutrientes (columnas) como llaves en cada diccionario de cada ingredientes
for key in ingr_dict: 
    new_dict2 = dict.fromkeys(nutrients, 0)
    ingr_dict[key] = new_dict2
#print(ingr_dict)
```

```python
#i = 0
#for key, value in ingr_dict.items():
#    for item1 in nutrients:
#        if i >= len(complete_value_list):
#            break
#        ingr_dict[key][item1] = complete_value_list[i]
#        i = i+1 
#print(ingr_dict)

for i in range(len(value_list)):
    for ii in range(len(value_list[0])):
        key = ingredient_list[i]
        item1 = nutrients[ii]
        ingr_dict[key][item1] = value_list[i][ii]
#print(ingr_dict)
```

```python
#nuevo diccionario principal para hacer comparación de valores
var_dict = {}
for ingr in ingredient_list: 
    var_dict[ingr] = {}

for key in var_dict: 
    new_dict3 = dict.fromkeys(nutrients, 0)
    var_dict[key] = new_dict3
```

```python
#función para elegir un ingrediente a comparar en el diccionario principal 
def compare_ing(dicc, ingr):
    var_dict2 = {}
    
    reference_values = dicc.get(ingr) 
    
    #estableciendo los ingredientes como llave principal del diccionario de variaciones
    for ing_compare in ingredient_list: 
        compare_values = dicc.get(ing_compare)
        var_dict2[ing_compare] = {}
        #estableciendo los nutrientes como llaves secundarias del diccionario de variaciones de cada ingrediente
        for key in compare_values.keys():
            nutr_compare = key
            if nutr_compare in reference_values: 
                #calculando las variaciones de cada nutriente 
                ref = reference_values.get(nutr_compare)
                val = compare_values.get(nutr_compare)
                if ref == 0 or ref == "nan": 
                    var = "nan"
                else: 
                    var = round(abs((ref - val)/ref), 3)
                var_dict2[ing_compare][nutr_compare] = var
                
    return(var_dict2)


trial = compare_ing(ingr_dict, "Salvado de trigo")

#printing like this so it can be visualized better
for key, value in trial.items(): 
    ingredient = key 
    nutr_dicc = value
    print(ingredient, nutr_dicc)
    
```
