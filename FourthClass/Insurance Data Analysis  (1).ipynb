##### Which gender is charged the highest rate?

```python
gender = insurance_data.groupby("sex")
mean = gender.mean()
mean
```

##### Which region has the most amount of smokers?

```python
insurance_data['smoker'] = insurance_data['smoker'] == "yes"
region = insurance_data.groupby("region")
region.sum()['smoker'].idxmax()
```

```python
insurance_data.pivot_table(index='region', aggfunc='sum')['smoker']
```

##### Who is getting charged 3 standard deviations more than the others in their age group?


```python
age_groups = insurance_data.groupby("age")
insurance_data["mean_charges"] = age_groups["charges"].transform(lambda x: x.mean())
insurance_data["std_charges"] = age_groups["charges"].transform(lambda x: x.std())
insurance_data
```

```python
age_groups["charges"].transform(lambda x: print(x.mean()))
```

```python
filter_1 = 3*(insurance_data["std_charges"]) > insurance_data["charges"]
insurance_data[filter_1]
```

```python
insurance_data.query("3*std_charges > charges")
```
